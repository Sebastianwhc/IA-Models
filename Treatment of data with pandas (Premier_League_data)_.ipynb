Best seller category, use of the function "group by"

```python
import pandas as pd
import numpy as np

#Para cargar el Dataset desde el drive
from google.colab import drive
drive.mount('/content/drive')
file_path = '/content/drive/MyDrive/bestsellers-with-categories.csv'
books = pd.read_csv(file_path)
books
```

Agrupación por autor y agregación por medio de calificaciones y reseñas de usuarios

```python
grouped = books.groupby('Author').agg({'User Rating': 'mean', 'Reviews': 'mean'})
print(grouped.head())
```

Agrupando por autor y género y contando el número de filas para cada grupo

```python
grouped = books.groupby(['Author', 'Genre']).count()

print(grouped.head())
```

Agrupando por Autor y sumando la columna Reseñas

```python
grouped = books.groupby('Author')['Reviews'].sum()
print(grouped.head())
```

Agrupación por autor y aplicación de diferentes funciones de agregación a diferentes columnas

```python
grouped = books.groupby('Author').agg({'Name': 'count', 'User Rating': 'mean', 'Reviews': 'sum'})
print(grouped.head())
```

Agrupación por autor y cálculo de la calificación de usuario mínima, máxima y mediana

```python
grouped = books.groupby('Author')['User Rating'].agg(['min', 'max', 'median'])
print(grouped.head())
```
